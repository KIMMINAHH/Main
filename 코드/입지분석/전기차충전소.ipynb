```python
import pandas as pd
import geopandas as gpd
```

```python
charger = gpd.read_file('../../데이터/입지분석/전기차충전소/시범 전기차충전소(시군구격자) 접근성.shp', encoding = 'utf-8')
charger = charger[charger['sido_nm_k'] == '서울특별시']
charger
```

```python
def polygon_to_coordinates(x): 
    lon, lat = x.exterior.xy 
    return [[x, y] for x, y in zip(lon, lat)]
```

```python
charger = charger.to_crs(epsg=4326)
charger['coordinates'] = charger['geometry'].apply(polygon_to_coordinates)
charger
```

```python
Main = pd.read_excel('../../데이터/입지분석/인구+주차장.xlsx', index_col=0)
charger = charger.set_index('gid')
charger
```

```python
charger = charger[['value']]
charger.columns = ['charger_access']
charger
```

```python
Main = Main.join(charger)
Main
```

```python
Main.to_excel('../../데이터/입지분석/인구+주차장+충전소.xlsx')
```

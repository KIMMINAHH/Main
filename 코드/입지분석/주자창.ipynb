```python
import pandas as pd
import geopandas as gpd
```

```python
parking = gpd.read_file('../../데이터/입지분석/주차장/176.2 주자창(시군구격자) 접근성.shp', encoding = 'utf-8')
parking.head()
```

필요한 데이터만 고르기

```python
parking = parking[['gid', 'sgg_nm_k', 'sido_nm_k', 'value', 'geometry']]

parking
```

```python
def polygon_to_coordinates(x): 
    lon, lat = x.exterior.xy 
    return [[x, y] for x, y in zip(lon, lat)]
```

```python
parking = parking.to_crs(epsg = 4326)
parking['coordinates'] = parking['geometry'].apply(polygon_to_coordinates)

parking
```

```python
parking = parking[parking['sido_nm_k'] == '서울특별시']
parking = parking.set_index('gid')
parking
```

```python
Main = pd.read_excel('../../데이터/입지분석/인구수_완.xlsx', index_col=1)
Main = Main.drop(['Unnamed: 0'], axis=1)
```

```python
Main = Main[['val', 'geometry', 'coordinates', 'coord_center', 'regularize_val']]
Main.columns = ['population', 'geometry', 'coordinates', 'coord_center', 'regularize_population']
```

```python
Main = Main.join(parking[['sgg_nm_k', 'sido_nm_k', 'value']])
```

```python
Main.columns = ['population', 'geometry', 'coordinates', 'coord_center', 'regularize_population', 'sgg_nm_k', 'sido_nm_k','parking_access']
Main
```
